```python
import pandas as pd
```

```python
df = pd.read_csv("/content/transformed_tweets.csv")
```

```python
df[df['groq_transformed']==0]
```

```python

# LLM status columns
llm_cols = {
    "Groq": "groq_transformed",
    "Gemini": "gemini_transformed",
    "GPT": "gpt_transformed",
    "DeepSeek": "deepseek_transformed"
}

# Convert to boolean safely
for model, col in llm_cols.items():
    df[col] = df[col].astype(bool)

# ---------------------------------------------------
# 1. Examples each LLM FAILED to transform
# ---------------------------------------------------

failed_examples = {}

for model, col in llm_cols.items():
    failed_df = df[df[col] == False]

    failed_examples[model] = failed_df

    print(f"\n{'='*50}")
    print(f"{model} failed on {len(failed_df)} examples")
    print(f"{'='*50}")

    print(
        failed_df[
            ['tweets', 'labels']
        ].head(10)
    )

# ---------------------------------------------------
# 2. Find COMMON failures across all LLMs
# ---------------------------------------------------

all_failed_mask = (
    (~df['groq_transformed']) &
    (~df['gemini_transformed']) &
    (~df['gpt_transformed']) &
    (~df['deepseek_transformed'])
)

common_failures = df[all_failed_mask]

print("\nCOMMON FAILURES ACROSS ALL LLMs")
print(common_failures[['tweets', 'labels']].head(20))

# ---------------------------------------------------
# 3. Find UNIQUE failures per LLM
# ---------------------------------------------------

groq_unique = df[
    (~df['groq_transformed']) &
    (df['gemini_transformed']) &
    (df['gpt_transformed']) &
    (df['deepseek_transformed'])
]

gemini_unique = df[
    (~df['gemini_transformed']) &
    (df['groq_transformed']) &
    (df['gpt_transformed']) &
    (df['deepseek_transformed'])
]

gpt_unique = df[
    (~df['gpt_transformed']) &
    (df['groq_transformed']) &
    (df['gemini_transformed']) &
    (df['deepseek_transformed'])
]

deepseek_unique = df[
    (~df['deepseek_transformed']) &
    (df['groq_transformed']) &
    (df['gemini_transformed']) &
    (df['gpt_transformed'])
]

print("\nGroq unique failures:", len(groq_unique))
print("Gemini unique failures:", len(gemini_unique))
print("GPT unique failures:", len(gpt_unique))
print("DeepSeek unique failures:", len(deepseek_unique))
```

```python
failure_pattern = pd.DataFrame({
    'Groq_failed': ~df['groq_transformed'],
    'Gemini_failed': ~df['gemini_transformed'],
    'GPT_failed': ~df['gpt_transformed'],
    'DeepSeek_failed': ~df['deepseek_transformed']
})

pattern_counts = (
    failure_pattern
    .value_counts()
    .reset_index(name='count')
)

print(pattern_counts)
```
